# Data visualization
The first step in the data analysis should always be a visual inspection of the rawdata.


```python
# Import standard packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns

# Generate the data
x = np.random.randn(500)
```

## Scatter Plots

```python
# Plot-command start ---------------------
plt.plot(x, '.')
# Plot-command end -----------------------
# Show plot
plt.show()
```

## Histogram

```python
plt.hist(x, bins=25)
```

## Cumulative Frequencies

```python
plt.plot(stats.cumfreq(x,numbins=25)[0])
```

## Box Plots

```python
plt.boxplot(x, sym='*')
```

## Grouped Bar Charts

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot(kind='bar', grid=False)
```

## Pie Charts

```python
import seaborn as sns
import matplotlib.pyplot as plt
txtLabels = 'Cats', 'Dogs', 'Frogs', 'Others'
fractions = [45, 30, 15, 10]
offsets =(0, 0.05, 0, 0)
plt.pie(fractions, explode=offsets, labels=txtLabels,autopct='%1.1f%%', shadow=True, startangle=90,
        colors=sns.color_palette('muted') )
plt.axis('equal')
```

## 3D Plots

```python
# imports specific to the plots in this example
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import get_test_data

# Twice as wide as it is tall.
fig = plt.figure(figsize=plt.figaspect(0.5))

#---- First subplot
# Note that the declaration "projection='3d'" is required for 3d plots!
ax = fig.add_subplot(1, 2, 1, projection='3d')

# Generate the grid
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)

# Generate the surface data
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
cmap=cm.GnBu, linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)


#---- Second subplot
ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y, Z = get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)



outfile = '3dGraph.png'
plt.savefig(outfile, dpi=200)
print('Image saved to {0}'.format(outfile))
plt.show()
```

## Correlation matrix

```python
# Create simulated datset in Python:
rs = np.random.RandomState(0)
df = pd.DataFrame(rs.rand(10, 10))
# Create and print correlation matrix:
corr = df.corr()
print(corr)

# Limit number of digits:
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```
