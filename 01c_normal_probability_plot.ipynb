## Q-Q plot, Normal probability plot

```python
from scipy.stats import norm
import numpy as np
fraction = np.linspace(0.05,0.95,num=19)
```

```python
vals1 = norm.ppf(fraction, 0, 1)    # percentile point function, inverse cdf
vals2 = norm.ppf(fraction, 2, 4)
                

import matplotlib.pyplot as plt
plt.scatter(vals1, vals2)
```

```python
import numpy as np 
import statsmodels.api as sm 
import pylab as py 
  

data_points = np.random.normal(0, 1, 100)     # mean=0, std=1, 100 random data points generated 
sm.qqplot(data_points, line ='45') 
py.show() 
```

```python
data_points = np.random.normal(2, 1, 100)     # mean =2, std = 1
sm.qqplot(data_points, line ='45') 
py.show() 
```

```python
data_points = np.random.uniform(1, 10, 100)     # low = 1, up=10, uniform
sm.qqplot(data_points, line ='45') 
py.show() 
```
