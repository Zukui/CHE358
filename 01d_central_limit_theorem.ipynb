### demonstration of central limit theorem: sample mean follows normal distribution as sample size increases


```python
import numpy as np
import random
import matplotlib.pyplot as plt

#s = np.random.gamma(2, 2, 100000)  # gamma distribution as population
s = np.random.randint(1, 7, size=100000)  # discrete uniform as population, e.g., throw dice 1-6
#s = np.random.chisquare(2,100000) # chi-square distribution as population

# Step 1
## sample with different sample size
# list of sample mean
meansample = []
# number of sampling
numofsample = 10000
# sample size
samplesize = [1,5,10,30]
# for each sample size 
for i in samplesize:
    eachmeansample = [] # collect mean of each sample
    # for each sampling
    for j in range(0,numofsample):
        rc = random.choices(s, k=i)  # sampling i samples from population
        eachmeansample.append(sum(rc)/len(rc))  # collect mean of each sample set
    # add mean of each sampling to the list
    meansample.append(eachmeansample)

# Step 2
# plot
cols = 2
rows = 2
fig, ax = plt.subplots(rows, cols, figsize=(20,15))
n = 0
for i in range(0, rows):
    for j in range(0, cols):
        ax[i, j].hist(meansample[n], 100, density=False)
        ax[i, j].set_title(label="sample size :" + str(samplesize[n]))
        n += 1
plt.show()
```
